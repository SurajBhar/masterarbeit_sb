```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_experiment_results(train_bal_acc_path, test_bal_acc_path, val_bal_acc_path,
                            train_loss_path, test_loss_path, val_loss_path,
                            lr_path, experiment_name):
    # Load the data from CSV files
    train_bal_acc = pd.read_csv(train_bal_acc_path)
    test_bal_acc = pd.read_csv(test_bal_acc_path)
    val_bal_acc = pd.read_csv(val_bal_acc_path)

    train_loss = pd.read_csv(train_loss_path)
    test_loss = pd.read_csv(test_loss_path)
    val_loss = pd.read_csv(val_loss_path)

    lr_data = pd.read_csv(lr_path)

    # Plot balanced accuracy for train, test, and validation
    plt.figure(figsize=(10, 6))
    plt.plot(train_bal_acc['Step'], train_bal_acc['Value']*100, label='Train Balanced Accuracy')
    plt.plot(test_bal_acc['Step'], test_bal_acc['Value']*100, label='Test Balanced Accuracy')
    plt.plot(val_bal_acc['Step'], val_bal_acc['Value']*100, label='Validation Balanced Accuracy')
    plt.title(f'Balanced Accuracy across Epochs for {experiment_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Balanced Accuracy (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot loss for train, test, and validation
    plt.figure(figsize=(10, 6))
    plt.plot(train_loss['Step'], train_loss['Value'], label='Train Loss')
    plt.plot(test_loss['Step'], test_loss['Value'], label='Test Loss')
    plt.plot(val_loss['Step'], val_loss['Value'], label='Validation Loss')
    plt.title(f'Loss across Epochs for {experiment_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot learning rate
    plt.figure(figsize=(10, 6))
    plt.plot(lr_data['Step'], lr_data['Value'], label='Learning Rate')
    plt.title('Learning Rate across Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
```
